# **5 different function of string**


# Slicing string

```
a="python language"
print(a[5:11])
```

# Concatenate String

```
a="Python"
b="Language"
print(a +" " +b)
```

# Replace String

```
a="best lucky"
print(a.replace("best","good"))
```

# Lower case function 

```
name="WELCOME TO PYTHON"
print(name.lower())
```

# Upper case funtion

```
name="welcome to python"
print(name.upper())
```

# **5 different function of the list object**

# Append list

```
no=[1,2,3,4,5,6,7,8]
no.append(9)
print(no)
```

# Clear List

```
list=["Better","good","Average"]
list.clear()
print(list)
```

# Reverse List

```
list1=["insert","delete","remove"]
list1.reverse()
print(list1)
```

# Sort List

```
list2=["B","D","Z","T","A"]
list2.sort()
print(list2)
```

# Count List

```
list3=["AA","BB","CC","AA","ZZ"]
x= list3.count("AA")
print(x)
```

# ***5 different functions of dictionary ***

# Get Dictionary

```
fruits={"fruit1":"Apple","fruit2":"Orange","fruit3":"Banana"}
print(fruits.get("fruit1"))
```

# Update Dictionary

```
fruits={"fruit1":"Apple","fruit2":"Orange","fruit3":"Banana"}
fruits.update({"fruit1":"Mango"})
print(fruits)
```

# Values Dictionary

```
person={"name":"Shraddha","age":25,"Address":"Pune"}
print(person.values())
```

# Keys Dictionary

```
person={"name":"Shraddha","age":25,"Address":"Pune"}
print(person.keys())
```

# Clear Dictionary

```
person={"name":"Shraddha","age":25,"Address":"Pune"}
print(person.clear())
```
